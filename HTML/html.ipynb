### 核心函数
select()- 用CSS selector找元素  

```python
from bs4 import BeautifulSoup
dom = BeautifulSoup("""
<html>
    <body>
        <h1>foo</h1>
        <p>bar
        <h1 class='even-headier'>baz</h1>
        <a href='https://www.nyu.edu/'>main page</a>
        <a href='https://cs.nyu.edu/home/index.html'>cs!</a>
        </p>
    </body>
</html>
""", "html.parser")
```

基本用法

```python
# get the elements with class .even-headier
print(dom.select('.even-headier'))
```

单个元素

```python
dom.select_one("h1")
```

```python
# get all h1's ... show the text of the first one
print(dom.select('h1'))
print(dom.select('h1')[0].get_text())
```

```python
# show the href attribute of the first link
print(dom.select('a')[0]['href'])
```

### Select可以选什么？

```python
dom.select("a")
```

- 2. 按照class选

```python
print(dom.select(".even-headier"))
```

- 3. 按id选

```python
dom.select("#main")
```

- 4. 组合选择

```python
dom.select("table tr td")
```

### 常见html标记
- a 链接
- h1 标题
- p 段落
- table 表格
- tr 行
- td 单元格

```python
import requests
from bs4 import BeautifulSoup

p = "https://en.wikipedia.org/wiki/List_of_sovereign_states"

headers = {
    "User-Agent": "Mozilla/5.0"
} # 告诉网站这是正常的浏览器访问，而不是爬虫

p2 = requests.get(p, headers=headers) # 发送HTTP GET请求，获取网页内容
dom = BeautifulSoup(p2.text, "html.parser") # 解析网页内容，创建一个BeautifulSoup对象

print(p2.status_code)# 打印HTTP响应状态码
print(len(dom.select("tr")))# 打印状态码和网页中所有表格行的数量

```

### HTML 里面的标签：
- 1. "a" 是一个tag 标签 全名叫超链接
- <a href ="https://google.com"Google</a>>
<a>标签名 href 属性名 Google 显示的文字

```python
link = dom.select_one("a")
print(link.text)    # 打印第一个链接的文本内容
print(link['href']) # 打印第一个链接的href属性值
```

### Relationship
a b       # 子孙（descendant）  
a > b     # 直接子（child）  

a.b ... element a with class name b (a class="b">...)  
a b ... all elements b that are descendants of a (regardless of level)  
a > b ... every b that is a child (direct descendant) of a  
a:nth-child(2) ... anchor tag that's the second child of its parent
