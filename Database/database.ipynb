### DQL (SELECT)

DQL Data Query Language

SELECT = 查数据


```python
SELECT col1, col2
FROM table
WHERE condition
ORDER BY col
LIMIT n;
```

1️⃣ FROM（数据从哪来）
FROM student

👉 从 student 表拿数据

🔹 2️⃣ WHERE（筛选）
WHERE score > 80

👉 只要分数 > 80 的人

🔹 3️⃣ SELECT（输出什么）
SELECT name, score

👉 只显示 name 和 score

🔹 4️⃣ ORDER BY（排序）
ORDER BY score DESC

👉 从高到低

🔹 5️⃣ LIMIT（限制数量）
LIMIT 5

👉 只要前5条

```python
-- Example:
SELECT title, budget, gross, 
ROUND(((gross - budget) / budget)::numeric, 2) AS roi
FROM movie
ORDER BY roi DESC;
```

1️⃣ SELECT
title, budget, gross

👉 直接选列

2️⃣ 计算 ROI
(gross - budget) / budget

👉 投资回报率

3️⃣ ::numeric

👉 转成 numeric（为了精确 + 可以 round）

4️⃣ ROUND(..., 2)

👉 保留2位小数

5️⃣ AS roi

👉 给这一列取名 roi

6️⃣ FROM
FROM movie

👉 从 movie 表拿数据

7️⃣ ORDER BY
ORDER BY roi DESC

👉 按 ROI 从大到小排序

```python
-- EXAMPLE

-- give me col1, col2, and new_col only
-- new_col is a calculated field
-- this is the 'SELECT list'
SELECT col1, col2, col3 * 2 as new_col    

	-- from table, some_Table
	FROM some_table

	-- the value in col1 must be > 1
	-- for the row to be returned
	WHERE col1 > 1

	-- sort by col2 ascending
	ORDER BY col2

	-- only give back, at most, 5 rows
	LIMIT 5;
```

- FROM to determine set of all possible rows to return!  
- WHERE to filter out rows that don't match criteria  
- SELECT list to determine the actual values of the resulting rows   by evaluating expressions, resolving column names, etc.  
- DISTINCT to eliminate duplicate rows in the output / DISTINCT ON  
- ORDER BY to sort the output rows  
- LIMIT to restrict the output rows to a specific number
